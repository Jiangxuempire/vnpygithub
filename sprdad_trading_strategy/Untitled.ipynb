```python
import importlib
from datetime import datetime

from vnpy.app.spread_trading.backtesting import BacktestingEngine
from vnpy.app.spread_trading.base import LegData, SpreadData
```

```python
# from strategies.statistical_arbitrage_grid_strategy import StatisticalArbitrageGridStrategy
from strategies.statistical_arbitrage_grid_strategy import StatisticalArbitrageGridStrategy
from strategies.statistical_arbitrage_strategy import StatisticalArbitrageStrategy
```

```python
spread = SpreadData(
    name="BTC-Spread",
    legs=[LegData("BTC-USD-B.OKEX"), LegData("BTC-USD-Q.OKEX")],
    price_multipliers={"BTC-USD-B.OKEX": 1, "BTC-USD-Q.OKEX": -1},
    trading_multipliers={"BTC-USD-B.OKEX": 1, "BTC-USD-Q.OKEX": -1},
    active_symbol="BTC-USD-B.OKEX",
    inverse_contracts={"BTC-USD-B.OKEX": True, "BTC-USD-Q.OKEX": False},
    min_volume=1
)
```

```python
# 计算手续费
rate_1 = 0.05/100
rate_2 = 0.05/100
rate = (rate_1 + rate_2) / 2
```

```python
engine = BacktestingEngine()
engine.set_parameters(
    spread=spread,
    interval="1m",
    start=datetime(2020, 3, 1),
    end=datetime(2020, 6, 12),
    rate=rate,
    slippage=0.1,  # 滑点
    size=100,
    pricetick=0.5,
    capital=100_000,
)
```

```python
statarb_setting = {
    "boll_window": 70,
    "boll_dev": 3.8,
    "fixed_pos":1.0,
}

statarb_setting1 = {
    "boll_window": 80,
    "boll_dev": 3,
    "fixed_pos":1.0,
    "max_pos" : 4,
    "price_proportion": 2.0,
    "pay_up":5,
    "increase_price": 10,       
    "interval": 5
}
```

```python
# engine.add_strategy(StatisticalArbitrageGridStrategy, statarb_setting1)
engine.add_strategy(StatisticalArbitrageStrategy, statarb_setting)
```

```python
engine.load_data()
engine.run_backtesting()
df = engine.calculate_result()
engine.calculate_statistics()
engine.show_chart()
```
