```python
import datetime
from dateutil.relativedelta import relativedelta
from vnpy.trader.database import database_manager
from vnpy.trader.constant import (Exchange,Interval)
months_1 = relativedelta(months=1)

start = datetime.date(2016,1,1)
product = "IF"
settle_dates = []
def gengerate_arb_dict(start, product):
    contracts = []
    
    end = datetime.datetime.now().date()
    dts = []
    while start < end:
        dts.append(start)
        start = start + months_1
        
    for i in dts:
        contract = product + i.strftime("%y%m")
        contracts.append(contract)
        data = database_manager.load_bar_data(
            symbol=contract, 
            exchange=Exchange.CFFEX, 
            interval=Interval.MINUTE, 
            start=i, 
            end= i + months_1)
        if not data:
            return
        settlment_time = data[-1].datetime.date()
        print(contract,"@",settlment_time)
        settle_dates.append(settlment_time)
```

```python
months_3 = relativedelta(months=3)
last = None
timec = {}
for dt in settle_dates:
    if last:
        dt_1 = dt
        p_1 = "IF" + dt_1.strftime("%y%m")
        
        dt_2 = dt + months_1
        p_2 = "IF" + dt_2.strftime("%y%m")
        
        b=((dt_2.month -1)//3 )*3+3
        a = dt_2.month

        v1 = b-a
        if v1 == 0:
            v = 3
        else:
            v = v1

        dt_3 = dt_2 + relativedelta(months=v)
        p_3 = "IF" + dt_3.strftime("%y%m")        
        dt_4 = dt_3 + months_3
        p_4 = "IF" + dt_4.strftime("%y%m")       
        timec[last] = [dt,p_1, p_2, p_3,p_4]
          
    last = dt + relativedelta(days=1)
```

```python
import importlib
from datetime import datetime

from vnpy.app.spread_trading.backtesting import BacktestingEngine
from vnpy.app.spread_trading.base import LegData, SpreadData

from statistical_arbitrage_strategy import StatisticalArbitrageStrategy
import grid_trading_strategy
importlib.reload(grid_trading_strategy)
from grid_trading_strategy import GridTradingStrategy
```

```python
def new_spread(act, pas):
    spread = SpreadData(
        name="IF-Spread",
        legs=[LegData(act), LegData(pas)],
        price_multipliers={act: 1, pas: -1},
        trading_multipliers={act: 1, pas: -1},
        active_symbol=act,
        inverse_contracts={act: False, pas: False},
        min_volume=1
    )

    return spread
```

```python
ll=[]
tradess = []
for k, v in timec.items():
    start = k
    end = v[0]
    start_history = start - relativedelta(days=20)
    print("start=", start, "end=", end, "start hist=",start_history)
    
    active = v[1]+ ".CFFEX"
    passive = v[4] + ".CFFEX"
    spread = new_spread(active, passive)
    print("active =", active, "passive=", passive)
    

    engine = BacktestingEngine()
    engine.set_parameters(
        spread=spread,
        interval="1m",
        start=start_history,
        end=end,
        rate=0.3/10000,
        slippage=0.2,
        size=300,
        pricetick=0.2,
        capital=100_000,
    )

    statarb_setting = {
        "boll_window": 60,
        "boll_dev": 4
    }

    grid_setting = {
        "grid_price": 20,
        "grid_volume": 1,
        "max_pos": 3
    }
    engine.add_strategy(GridTradingStrategy, grid_setting)
    engine.load_data()

    engine.run_backtesting()
    n = engine.calculate_result()
    engine.calculate_statistics()
    engine.show_chart()
    # if n.any():
    if n is not None:
        m = n.loc[start:end, :]
        ll.append(m)
        
    tt = engine.trades
    print("@", tt)
    if tt:
        for k, v in tt.items():
            da = v.datetime.date()
            print("@", da, start, end)
            if da >= start and da <= end:
                tradess.append(v)
                
    #tradess.append(tt)
```
